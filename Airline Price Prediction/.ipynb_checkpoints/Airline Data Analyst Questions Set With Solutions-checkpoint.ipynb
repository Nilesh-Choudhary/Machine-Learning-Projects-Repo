```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
data = pd.read_excel("Data_Train.xlsx")
```

```python
data.head()
```

# Q1) Find the average price of non-stop flights and flights with stops. Create a bar chart to compare the average prices of these two types of flights.

```python
non_stop_avg_price = data[data['Total_Stops'] == 'non-stop']['Price'].mean()

stops_avg_price = data[data['Total_Stops'] != 'non-stop']['Price'].mean()

print(f"Average price for non-stop flights: {non_stop_avg_price}")
print(f"Average price for flights with stops: {stops_avg_price}")

avg_prices = [non_stop_avg_price, stops_avg_price]
labels = ['Non-stop Flights', 'Flights with Stops']

plt.figure(figsize=(8, 5))
plt.bar(labels, avg_prices, color=['green', 'blue'])
plt.title('Average Price Comparison: Non-stop vs Flights with Stops')
plt.ylabel('Average Price')
plt.show()
```

# Q2) Find the flight with the longest duration and display its details.

```python
def convert_duration_to_hours(duration):
    hours = 0
    if 'h' in duration:
        parts = duration.split('h')
        hours = int(parts[0].strip())  # Extract hours
    return hours

data['Duration_Hours'] = data['Duration'].apply(convert_duration_to_hours)

longest_duration_flight = data.loc[data['Duration_Hours'].idxmax()]
print(longest_duration_flight)
```

# Q3) What are the top 5 most frequent routes among all the flights?

```python
top_5_routes = data['Route'].value_counts().head(5)
print(top_5_routes)

plt.figure(figsize=(10, 6))
top_5_routes.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Top 5 Most Frequent Routes')
plt.xlabel('Route')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

# Q4) How many flights are available on each day of the week? Create a bar chart to show the number of flights for each day.

```python
from datetime import datetime

data['Date_of_Journey'] = pd.to_datetime(data['Date_of_Journey'], format='%d/%m/%Y')

data['Day_of_Week'] = data['Date_of_Journey'].dt.day_name()

flights_per_day = data['Day_of_Week'].value_counts()

print(flights_per_day)

plt.figure(figsize=(10, 6))
flights_per_day.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Flights per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45)
plt.show()
```

# Q5) Find the source and destination pairs with more than 1000 flights. Display the result in a table and create a bar chart to show the number of flights for each popular source-destination pair.

```python
source_destination_counts = data.groupby(['Source', 'Destination']).size().reset_index(name='Flight_Count')

popular_routes = source_destination_counts[source_destination_counts['Flight_Count'] > 1000]
print(popular_routes)

plt.figure(figsize=(12, 6))
plt.barh(popular_routes['Source'] + ' → ' + popular_routes['Destination'], popular_routes['Flight_Count'], color='skyblue')
plt.title('Popular Source-Destination Pairs with More Than 1000 Flights')
plt.xlabel('Number of Flights')
plt.ylabel('Source → Destination')
plt.show()
```

# Q6) Categorize the flights based on their departure time into Morning, Afternoon, Evening, and Night. Create a new column that indicates whether a flight departs in the morning (5 AM to 12 PM), afternoon (12 PM to 5 PM), evening (5 PM to 9 PM), or night (9 PM to 5 AM). Create a bar chart to show how many flights depart in each of these time categories.

```python
def time_of_day(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

data['Dep_Hour'] = pd.to_datetime(data['Dep_Time']).dt.hour

data['Time_of_Day'] = data['Dep_Hour'].apply(time_of_day)

time_of_day_counts = data['Time_of_Day'].value_counts()

print(data[['Dep_Time', 'Time_of_Day']].head())

plt.figure(figsize=(8, 6))
time_of_day_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Flights by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45)
plt.show()
```

# Q7) Calculate the average flight price for each number of stops and visualize the results in a bar chart?

```python
avg_price_by_stops = data.groupby('Total_Stops')['Price'].mean()

print(avg_price_by_stops)

plt.figure(figsize=(10, 6))
avg_price_by_stops.plot(kind='bar', color='coral')
plt.title('Average Flight Price by Number of Stops')
plt.xlabel('Number of Stops')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

# Q8) Create a line chart showing the trend of average flight prices over each day of the month.

```python
data['Day'] = pd.to_datetime(data['Date_of_Journey']).dt.day

avg_price_by_day = data.groupby('Day')['Price'].mean()

plt.figure(figsize=(10, 6))
avg_price_by_day.plot(kind='line', marker='o', color='green')
plt.title('Average Flight Price Trend Over Days of the Month')
plt.xlabel('Day of the Month')
plt.ylabel('Average Price')
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.show()
```

# Q9) Create a horizontal bar chart showing the top 5 routes with the highest average flight price.

```python
data['Route'] = data['Source'] + ' to ' + data['Destination']

avg_price_by_route = data.groupby('Route')['Price'].mean().sort_values(ascending=False).head(5)

plt.figure(figsize=(10, 6))
avg_price_by_route.plot(kind='barh', color='purple')
plt.title('Top 5 Routes by Average Flight Price')
plt.xlabel('Average Price')
plt.ylabel('Route')
plt.show()
```

# Q10) Generate a violin plot to compare flight price distributions across the top 4 airlines.

```python
top_airlines = data['Airline'].value_counts().head(4).index

filtered_data = data[data['Airline'].isin(top_airlines)]

plt.figure(figsize=(10, 6))
sns.violinplot(x='Airline', y='Price', data=filtered_data)
plt.title('Flight Price Distribution for Top 4 Airlines')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.show()
```

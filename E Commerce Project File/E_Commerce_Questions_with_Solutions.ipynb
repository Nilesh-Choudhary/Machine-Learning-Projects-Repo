```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv("Phone_Data.csv")
```

```python
df.head()
```

# Q1) Create a new column called phone_brand in the DataFrame by extracting the brand name from the Phone_Name column. 

```python
df['phone_brand'] = df['Phone_Name'].str.split(' ').str[0]

df
```

# Q2) Calculate the number of unique phone brands in the dataset and generate a list of those unique phone brands, displaying both the count and the list.

```python
unique_brands = df['phone_brand'].nunique()

unique_brand_list = df['phone_brand'].unique()

print(f"Number of unique phone brands: {unique_brands}")
print(f"List of unique phone brands: {unique_brand_list}")
```

# Q3) Identify the phones with a 7000 mAh battery in the dataset, count them, display the ones priced over 20,000, and list the unique brands offering these phones.

```python
phones_with_large_battery = df[df['Battery'].str.contains('7000 mAh')]

print("Phones with large battery (7000 mAh):")
print(phones_with_large_battery)

count_large_battery_phones = phones_with_large_battery.shape[0]
print(f"Total number of phones with a 7000 mAh battery: {count_large_battery_phones}")

expensive_large_battery_phones = phones_with_large_battery[phones_with_large_battery['Price'] > 20000]
print("\nExpensive phones with large battery (Price > 20000):")
print(expensive_large_battery_phones[['Phone_Name', 'Battery', 'Price']])

unique_brands_large_battery = phones_with_large_battery['phone_brand'].unique()
print("\nUnique phone brands with large batteries:")
print(unique_brands_large_battery)
```

# Q4) Find phones that have Price greater than the average price for their phone_brand.

```python
avg_price_per_brand = df.groupby('phone_brand')['Price'].transform('mean')
phones_above_avg_price = df[df['Price'] > avg_price_per_brand]
phones_above_avg_price
```

# Q5) Categorize phones into 'Budget', 'Mid-Range', 'Premium', and 'High-End' based on their price. Count the number of phones in each category and display the results in a bar chart.

```python
conditions = [
    (df['Price'] <= 20000),  
    (df['Price'] > 20000) & (df['Price'] <= 50000),  
    (df['Price'] > 50000) & (df['Price'] <= 80000), 
    (df['Price'] > 80000)  
]

categories = ['Budget', 'Mid-Range', 'Premium', 'High-End']

df['Phone_Category'] = pd.cut(df['Price'], bins=[0, 20000, 50000, 80000, float('inf')], labels=categories, right=False)

category_counts = df['Phone_Category'].value_counts()

print(df)

print("\nPhone Category Counts:")
print(category_counts)

category_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Phone Categories by Price Range')
plt.xlabel('Phone Category')
plt.ylabel('Number of Phones')
plt.xticks(rotation=0)
plt.show()
```

# Q6) Create a bar chart showing the average Price per phone_brand.


```python
import matplotlib.pyplot as plt

avg_price_by_brand = df.groupby('phone_brand')['Price'].mean().sort_values()
avg_price_by_brand.plot(kind='bar', figsize=(10, 6), title='Average Price per Brand')
plt.xlabel('Phone Brand')
plt.ylabel('Average Price')
plt.show()
```

# Q7)  Create a histogram showing the distribution of Battery capacities.

```python
df['Battery_Capacity'] = df['Battery'].str.extract(r'(\d+)').astype(float)
df['Battery_Capacity'].hist(bins=10, figsize=(8, 5))
plt.title('Distribution of Battery Capacities')
plt.xlabel('Battery Capacity (mAh)')
plt.ylabel('Frequency')
plt.show()
```

# Q8) Determine the top 3 most common phone brands and visualize their frequency using a pie chart.

```python
brand_counts = df['phone_brand'].value_counts().head(3)

plt.figure(figsize=(6, 6))
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Top 3 Most Common Phone Brands')
plt.show()
```

# Q9) Identify and visualize the most common display sizes in the dataset using a count plot.

```python
top_5_display_sizes = df['Display'].value_counts().head(5).index

plt.figure(figsize=(10, 6))
sns.countplot(data=df[df['Display'].isin(top_5_display_sizes)], 
              y='Display', 
              order=top_5_display_sizes, 
              palette='Blues_r')
plt.title('Top 5 Most Common Display Sizes')
plt.xlabel('Count')
plt.ylabel('Display')
plt.show()
```

# Q10) Compare the average price of phones with different types of displays (e.g., HD, Full HD, Super Retina) using a bar chart.

```python
df['Display_Type'] = df['Display'].apply(lambda x: 'Super Retina' if 'Super Retina' in x else ('Full HD' if 'Full HD' in x else 'HD'))

avg_price_display_type = df.groupby('Display_Type')['Price'].mean()

plt.figure(figsize=(8, 5))
avg_price_display_type.plot(kind='bar', color='plum')
plt.title('Average Price by Display Type')
plt.xlabel('Display Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```
